#Encoding categorical variables

```python
df['Has Table booking'] = df['Has Table booking'].replace({'Yes': 1, 'No': 0})
df['Has Online delivery'] = df['Has Online delivery'].replace({'Yes': 1, 'No': 0})
df['Is delivering now'] = df['Is delivering now'].replace({'Yes': 1, 'No': 0})
df['Switch to order menu'] = df['Switch to order menu'].replace({'Yes': 1, 'No': 0})
df['Rating text'] = df['Rating text'].replace({'Poor': 0, 'Not rated': 1,'Average':2,'Good':3,'Very Good':4,'Excellent':5 })
df['Rating color'] = df['Rating color'].replace({'Red': 0, 'White': 1,'Orange':2,'Yellow':3,'Green':4,'Dark Green':5 })
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Assuming 'Cuisines' contains comma-separated lists of cuisines, e.g., "Italian, Chinese"
# First, split the 'Cuisines' column into a list of cuisines
df['Cuisines'] = df['Cuisines'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])

# Initialize the MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Apply MultiLabelBinarizer to create binary columns for each cuisine
cuisine_encoded = mlb.fit_transform(df['Cuisines'])

# Convert the binary matrix to a DataFrame with the cuisine names as column names
cuisine_encoded_df = pd.DataFrame(cuisine_encoded, columns=mlb.classes_)

# Add the new binary columns back to the original dataframe
df = pd.concat([df, cuisine_encoded_df], axis=1)

```

Determining the criteria for restaurant recommendations e.g., cuisine preference,price range.



*   Cuisine preference: The user could specify their preferred cuisines (e.g., Italian, Chinese, etc.).
*   Price range: The user could specify a price range (e.g., low, medium, high).
*   Rating: The user could prefer restaurants with higher ratings.
*   Delivery options: Whether the user wants delivery options or not.

```python
# List of features to use in the recommendation system
features = ['Average Cost for two', 'Price range', 'Aggregate rating', 'Has Table booking',
            'Has Online delivery', 'Is delivering now', 'Switch to order menu'] + list(cuisine_encoded_df.columns)

# Create the feature matrix X
X = df[features]
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize the numerical features (e.g., 'Average Cost for two', 'Price range', etc.)
numerical_features = ['Average Cost for two', 'Price range', 'Aggregate rating']
X[numerical_features] = StandardScaler().fit_transform(X[numerical_features])
```

#Implement Content-Based Filtering

Calculate Similarities

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Normalize numerical features
scaler = StandardScaler()
numerical_features = ['Average Cost for two', 'Price range', 'Aggregate rating']
df[numerical_features] = scaler.fit_transform(df[numerical_features])



# Compute cosine similarity between all restaurants
cosine_sim = cosine_similarity(X)

# Function to recommend restaurants based on a restaurant ID
def recommend_restaurants(restaurant_id, top_n=10):
    # Get the similarity scores for the restaurant
    sim_scores = list(enumerate(cosine_sim[restaurant_id]))

    # Sort the restaurants by similarity score (in descending order)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the top n most similar restaurants (excluding the restaurant itself)
    sim_scores = sim_scores[1:top_n+1]

    # Get restaurant indices
    restaurant_indices = [x[0] for x in sim_scores]

    # Return the most similar restaurants
    return df.iloc[restaurant_indices][['Restaurant ID', 'Cuisines', 'Average Cost for two', 'Price range', 'Aggregate rating']]

# Example: Recommend restaurants similar to the restaurant with ID = 5
recommended_restaurants = recommend_restaurants(restaurant_id=5, top_n=10)
print(recommended_restaurants)
```

Handling Cuisine Preferences

```python
def recommend_by_cuisine(cuisine_preference, top_n=10):
    # Filter the dataset by cuisine preference using multi-hot encoding
    # Check if the cuisine exists in the multi-hot encoded features
    filtered_data = df[df[cuisine_preference] == 1]

    # If no matching restaurants, return an empty dataframe or an alternative recommendation
    if filtered_data.empty:
        return pd.DataFrame()  # Return an empty DataFrame if no restaurants are found

    # Compute cosine similarity for the filtered restaurants
    X_filtered = filtered_data[features]
    cosine_sim_filtered = cosine_similarity(X_filtered)

    # Recommend top N similar restaurants based on the filtered data
    # We calculate cosine similarity for all restaurants, but we only use the filtered ones for recommendations
    sim_scores = list(enumerate(cosine_sim_filtered[0]))  # For the first restaurant in the filtered set
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]  # Get top N similar ones

    # Get the indices of the recommended restaurants
    recommended_indices = [x[0] for x in sim_scores]

    # Return the most similar restaurants from the filtered data
    return filtered_data.iloc[recommended_indices][['Restaurant ID', 'Cuisines', 'Average Cost for two', 'Price range', 'Aggregate rating']]

# Example: Recommend restaurants with cuisine preference 'Italian' (assuming 'Italian' corresponds to a binary column)
# Assuming 'Italian' is a column in the one-hot encoded data, for example 'French', 'Italian', 'Chinese' etc.
recommended_italian_restaurants = recommend_by_cuisine(cuisine_preference='Italian', top_n=5)
print(recommended_italian_restaurants)
```

#Test the Recommendation System
